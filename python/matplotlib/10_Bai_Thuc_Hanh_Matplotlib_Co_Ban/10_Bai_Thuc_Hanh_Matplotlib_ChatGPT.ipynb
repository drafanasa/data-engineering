üìò B√ÄI 5: Bi·ªÉu ƒë·ªì c·ªôt nh√≥m (Grouped Bar Chart)
Ch·ªß ƒë·ªÅ: So s√°nh doanh s·ªë gi·ªØa c√°c s·∫£n ph·∫©m theo t·ª´ng qu√Ω
B·ªëi c·∫£nh: Ban gi√°m ƒë·ªëc c·∫ßn xem xu h∆∞·ªõng theo th·ªùi gian.
M·ª•c ti√™u: V·∫Ω bi·ªÉu ƒë·ªì c·ªôt nh√≥m cho t·ª´ng qu√Ω.
C·ªôt d·ªØ li·ªáu:

product_name, quarter, units_sold

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sales_df = pd.read_csv('dataset_bai_5_doanh_so_theo_quy_va_san_pham_utf8.csv')
grouped_quarter_sale = sales_df.groupby(['qu√Ω','s·∫£n_ph·∫©m']).sum().reset_index()

# B∆∞·ªõc 1 - Pivot b·∫£ng
df_pivot = grouped_quarter_sale.pivot(index='qu√Ω', columns='s·∫£n_ph·∫©m', values='s·ªë_l∆∞·ª£ng_b√°n')

# B∆∞·ªõc 2 - V·∫Ω bi·ªÉu ƒë·ªì c·ªôt nh√≥m
labels = df_pivot.index.tolist()       # ['Q1', 'Q2', ...]
products = df_pivot.columns.tolist()   # ['√Åo thun', 'Qu·∫ßn jean', ...]

x = np.arange(len(labels))             # v·ªã tr√≠ tr·ª•c x
width = 0.15                           # ƒë·ªô r·ªông m·ªói c·ªôt

plt.figure(figsize=(10, 5))

for i, product in enumerate(products):
    plt.bar(x + i * width, df_pivot[product], width, label=product) 
    # x + i * width l√† c√°ch "ƒë·∫∑t l·ªách" t·ª´ng s·∫£n ph·∫©m trong c√πng m·ªôt nh√≥m (qu√Ω) -> gi√∫p d·ªãch c·ªôt sang ph·∫£i m·ªôt ƒë∆°n v·ªã ƒë·ªÉ t·∫°o "c·ªôt nh√≥m" cho m·ªói qu√Ω
    # df_pivot[product] l√† t√™n m·ªói s·∫£n ph·∫©m'√Åo thun', 'Qu·∫ßn jean', ... ƒë∆∞·ª£c m·ªói c·ªôt th·ªÉ hi·ªán
    # width l√† ƒë·ªô r·ªông m·ªói c·ªôt
    # label=product l√† t√™n s·∫Ω hi·ªÉn th·ªã trong ch√∫ th√≠ch (legend)

plt.xticks(x + width * (len(products)-1)/2, labels)
plt.xlabel("Qu√Ω")
plt.ylabel("S·ªë l∆∞·ª£ng b√°n")
plt.title("Doanh s·ªë s·∫£n ph·∫©m theo qu√Ω")
plt.legend(title="S·∫£n ph·∫©m", bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

üìò B√ÄI 6: Bi·ªÉu ƒë·ªì h·ªôp (Box Plot)
Ch·ªß ƒë·ªÅ: Ph√¢n ph·ªëi l∆∞∆°ng c·ªßa nh√¢n vi√™n theo ph√≤ng ban
B·ªëi c·∫£nh: Ph√≤ng nh√¢n s·ª± mu·ªën ki·ªÉm tra ch√™nh l·ªách l∆∞∆°ng gi·ªØa c√°c ph√≤ng.
M·ª•c ti√™u: V·∫Ω bi·ªÉu ƒë·ªì boxplot theo department.
C·ªôt d·ªØ li·ªáu:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("dataset_bai_6_luong_nhan_vien_theo_phong_ban_utf8.csv")
#print(df.head(30))

# Nh√≥m l∆∞∆°ng theo ph√≤ng ban th√†nh m·ªôt DANH S√ÅCH 'groups'
groups = []
for dept in df['ph√≤ng_ban'].unique():
    mask = df[df['ph√≤ng_ban'] == dept]  #  boolean indexing / boolean masking l·ªçc c√°c d√≤ng thu·ªôc ph√≤ng ban hi·ªán t·∫°i
    values = mask['l∆∞∆°ng_th√°ng_tri·ªáu_ƒë·ªìng'].values  # l·∫•y danh s√°ch l∆∞∆°ng
    groups.append(values)  # th√™m v√†o danh s√°ch t·ªïng
#print(groups)

plt.figure(figsize=(10, 5))
plt.boxplot(groups, tick_labels=df['ph√≤ng_ban'].unique())
plt.title("So s√°nh ph√¢n ph·ªëi l∆∞∆°ng nh√¢n vi√™n theo ph√≤ng ban")
plt.xlabel("Ph√≤ng ban")
plt.ylabel("L∆∞∆°ng th√°ng (tri·ªáu ƒë·ªìng)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

üìò B√ÄI 8: Bi·ªÉu ƒë·ªì c·ªôt x·∫øp ch·ªìng (Stacked Bar Chart)
Ch·ªß ƒë·ªÅ: Churn Rate theo th√°ng v√† lo·∫°i t√†i kho·∫£n
B·ªëi c·∫£nh: C√¥ng ty SaaS mu·ªën xem lo·∫°i t√†i kho·∫£n n√†o churn nhi·ªÅu.
M·ª•c ti√™u: V·∫Ω bi·ªÉu ƒë·ªì c·ªôt x·∫øp ch·ªìng theo month v√† account_type.
C·ªôt d·ªØ li·ªáu:

month, account_type, churned_users

```python
import pandas as pd
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu
df = pd.read_csv("dataset_bai_8_churn_theo_thang_va_loai_tai_khoan_utf8.csv")

# Pivot d·ªØ li·ªáu: rows = month, columns = account_type, values = churned_users
pivot_df = df.pivot(index='month', columns='account_type', values='churned_users').fillna(0)
print(pivot_df)

# T√≠nh t·ªïng t·ª´ng c·ªôt ƒë·ªÉ chu·∫©n b·ªã t√≠nh ph·∫ßn trƒÉm
total = pivot_df.sum(axis=1)

# T√≠nh ph·∫ßn trƒÉm ƒë√≥ng g√≥p
percent_df = pivot_df.divide(total, axis=0) * 100

# T·∫°o tr·ª•c x
months = pivot_df.index.tolist()
x = range(len(months))

# T·∫°o c√°c l·ªõp d·ªØ li·ªáu x·∫øp ch·ªìng
free = pivot_df['Free']
basic = pivot_df['Basic']
premium = pivot_df['Premium']

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(10, 5))
plt.bar(x, free, label='Free')
plt.bar(x, basic, bottom=free, label='Basic')
plt.bar(x, premium, bottom=free + basic, label='Premium')

# Th√™m label ph·∫ßn trƒÉm
for i in range(len(x)):
    # Label cho l·ªõp Free
    plt.text(x[i], free.iloc[i]/2,
             f"{percent_df.loc[percent_df.index[i], 'Free']:.1f}%",
             ha='center', va='center', color='white', fontsize=8)

    # Label cho l·ªõp Basic
    plt.text(x[i], free.iloc[i] + basic.iloc[i]/2,
             f"{percent_df.loc[percent_df.index[i], 'Basic']:.1f}%",
             ha='center', va='center', color='white', fontsize=8)

    # Label cho l·ªõp Premium
    plt.text(x[i], free.iloc[i] + basic.iloc[i] + premium.iloc[i]/2,
             f"{percent_df.loc[percent_df.index[i], 'Premium']:.1f}%",
             ha='center', va='center', color='white', fontsize=8)

# C√†i ƒë·∫∑t
plt.xticks(x, months)
plt.title("Churn Rate theo th√°ng v√† lo·∫°i t√†i kho·∫£n (% ƒë√≥ng g√≥p)")
plt.xlabel("Th√°ng")
plt.ylabel("S·ªë l∆∞·ª£ng ng∆∞·ªùi d√πng r·ªùi b·ªè")
plt.legend(title="Lo·∫°i t√†i kho·∫£n")
plt.tight_layout()
plt.show()
```

üìò B√ÄI 10: Bi·ªÉu ƒë·ªì heatmap (Ma tr·∫≠n nhi·ªát)
Ch·ªß ƒë·ªÅ: L∆∞·ª£ng ƒë∆°n h√†ng theo gi·ªù trong ng√†y v√† ng√†y trong tu·∫ßn
B·ªëi c·∫£nh: Nh√≥m v·∫≠n h√†nh mu·ªën t·ªëi ∆∞u th·ªùi gian giao h√†ng.
M·ª•c ti√™u: V·∫Ω heatmap gi·ªù √ó ng√†y trong tu·∫ßn.
C·ªôt d·ªØ li·ªáu:

day_of_week, hour, order_count

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu t·ª´ file CSV
df = pd.read_csv("dataset_bai_10_don_hang_theo_gio_va_thu_utf8.csv")

# Pivot b·∫£ng: t·ª´ long format ‚Üí wide format (d√≤ng = gi·ªù, c·ªôt = th·ª© trong tu·∫ßn)
# M·ªói √¥ l√† s·ªë ƒë∆°n h√†ng t·∫°i gi·ªù ƒë√≥, ng√†y ƒë√≥
pivot_table = df.pivot(index='hour', columns='day_of_week', values='order_count')

# ƒê·∫£m b·∫£o th·ª© trong tu·∫ßn hi·ªÉn th·ªã ƒë√∫ng th·ª© t·ª± (Mon ‚Üí Sun)
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
pivot_table = pivot_table[day_order]

# V·∫Ω heatmap
plt.figure(figsize=(10, 8))  # K√≠ch th∆∞·ªõc bi·ªÉu ƒë·ªì
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlOrRd')  # annot=True ƒë·ªÉ hi·ªán s·ªë, fmt='d' ƒë·ªãnh d·∫°ng s·ªë nguy√™n

# Tu·ª≥ ch·ªânh bi·ªÉu ƒë·ªì
plt.title("L∆∞·ª£ng ƒë∆°n h√†ng theo gi·ªù v√† ng√†y trong tu·∫ßn", fontsize=14)
plt.xlabel("Th·ª© trong tu·∫ßn")
plt.ylabel("Gi·ªù trong ng√†y")
plt.yticks(rotation=0)  # Gi·ªØ s·ªë gi·ªù n·∫±m ngang
plt.tight_layout()

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
```

```python
# Install các thư viện bằng CMD: pip install beautifulsoup4 requests rich
# Import requests và BeautifulSoup
import requests
from bs4 import BeautifulSoup
# Import rich để hiển thị HTML cực đẹp
from rich import print
from rich.syntax import Syntax
```

```python
# Gửi request đầu tiên đến trang web cần scrap data
book_url = 'https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html'
response = requests.get(book_url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Hiển trị HTML được làm đẹp bằng .prettify() và rich
rich_html = Syntax(soup.prettify(), "html", theme="solarized-dark", line_numbers=True)
print(rich_html)
```

```python
# Scraping lần lượt các dữ liệu về: name, price, category, stars, upc, availability, in_stock, image_link
```

```python
# Tìm và in nội dung text của kết quả đầu tiên trong dữ liệu 'name'
name = soup.find('div', class_='product_main').h1.text
print(name)
```

```python
# Tìm và in nội dung text của kết quả đầu tiên trong dữ liệu 'price'
price = soup.find('div', class_='product_main').p.text
print(price)
```

```python
# Tìm và in một ResultSet chứa tất cả kết quả của dữ liệu 'category'
ul_container = soup.find('ul', class_='breadcrumb')
li_items = ul_container.find_all('li')
print(f"ResultSet of 'category':\n{li_items}")
print(f'datatype: {type(li_items)}')

# Tìm và in nội dung của kết quả có index [2] trong dữ liệu 'category'
category = li_items[2].a.text
print(category)
```

```python
# Tìm vị trí của phần tử chứa dữ liệu về 'stars' (kết quả đầu tiên)
star_p_element = soup.find('p', class_='star-rating')

# Lấy danh sách các thuộc tính 'class' của phần tử vừa tìm được - ['star-rating', 'Three']
star_class_name_list = star_p_element['class']

# Lấy riêng dữ liệu về 'stars' từ danh sách trên (ở vị trí index [1])
star_string = star_class_name_list[1] # 'Three'

# Tạo từ dictionary chứa quy ước số stars để chuyển đổi string 'Three' thành number '3' tương ứng
number_dict = {'One': '1', 'Two': '2', 'Three': '3', 'Four': '4', 'Five': '5'}
stars = number_dict[star_string]

# In kết quả là số 'star' để sử dụng cho phân tích data sau này
print(stars)
```

```python
# Xác định vị trí của thẻ <th> chứa nội dung là 'UPC' (*UPC = Universal Product Code - Mã vạch sản phẩm)
upc_th = soup.find('th', string='UPC')

# Lấy và in nội dung text của mã UPC (theo HTML thì nó nằm ngay dưới thẻ <th>UPC</th>)
upc = upc_th.find_next_sibling().text 
print(upc)
```

```python
# Xác định vị trí của thẻ <th> chứa nội dung là 'Availability' (tương tự như lấy UPC ở ví dụ trên)
availability_th = soup.find('th', string='Availability')

# Lấy và in nội dung text của 'Availability'
availability = availability_th.find_next_sibling().text
print(availability)
```

```python
# Kết quả trước: In stock (22 available)
# Dùng hàm .split để tách chuỗi và in ra chỉ giá trị cần thiết là số '22' để sử dụng cho phân tích data sau này
try:
    in_stock = availability.split('(')[1].split(' ')[0]
except IndexError:
    in_stock = '0'  # Nếu không tìm thấy, mặc định là 0
print(in_stock)
```

```python
# Lấy đường dẫn của ảnh từ website

# Tìm vị trí chứa ảnh sản phẩm, sau đó lấy đường dẫn đến ảnh 'src' (path tương đối)
image_src = soup.find('div', class_='thumbnail').img['src'] # Output: '../../media/cache/fe/72/fe72f0532301ec28892ae79a629a293c.jpg'

# Loại bỏ 6 ký tự đầu ('../../') của path vừa lấy được
image_src_clean = image_src[6:]

# Nối path tương đối với URL gốc để tạo thành link tuyệt đối
image_link = 'https://books.toscrape.com/' + image_src_clean

# In ra link ảnh đầy đủ
print(image_link)
```
